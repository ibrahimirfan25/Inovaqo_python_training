```python
# assigning values
name = "Ibrahim"
age = 20
height = 5.8

print(name)
print(age)
print(height)

# multiple assignment
a, b, c = 10, 20, 30
print(a, b, c)

# constant convention
PI = 3.14159
print(PI)
```

```python
# Python private variable concept
public_var = "I am public"
_private_var = "I am 'private' by naming convention"
__very_private = "I am very private"

print(public_var)
print(_private_var)
print(__very_private)
```

```python
integer_num = 10
float_num = 10.5
complex_num = 2 + 3j

string_val = "Python"

list_val = [1, 2, 3]
tuple_val = (1, 2, 3)
range_val = range(5)

bytes_val = b"Hello"
bytearray_val = bytearray(b"Hello")

dict_val = {"name":"Ibrahim","age":20}
set_val = {1,2,3}

bool_val = True
none_val = None

print(type(integer_num))
print(type(float_num))
print(type(complex_num))
print(type(string_val))
print(type(list_val))
print(type(tuple_val))
print(type(range_val))
print(type(bytes_val))
print(type(bytearray_val))
print(type(dict_val))
print(type(set_val))
print(type(bool_val))
print(type(none_val))
```

```python
#type casting
x = "10"
y = "5.5"

int_val = int(x)
float_val = float(y)
str_val = str(100)

print(int_val)
print(float_val)
print(str_val)
```

```python
#Match case
day = 3

match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case _:
        print("Invalid Day")
```
