```python
# WHILE LOOPS
i = 1
while i <= 5:
    print(i)
    i += 1

i = 1
while i <= 5:
    if i == 3:
        break
    print(i)
    i += 1

i = 0
while i < 5:
    i += 1
    if i == 3:
        continue
    print(i)


# FOR LOOPS
for i in range(1, 6):
    print(i)

for ch in "python":
    print(ch)

nums = [10, 20, 30]
for index, value in enumerate(nums):
    print(index, value)

student = {"name": "Ibrahim", "age": 20, "course": "Python"}
for key, value in student.items():
    print(key, value)

for i in range(3):
    for j in range(2):
        print(i, j)


# FUNCTIONS
def greet():
    print("Hello!")


def greet_person(name):
    print("Hello", name)


def add(a, b):
    return a + b


def welcome(name="User"):
    print("Welcome", name)


def student_info(name, age):
    print(name, age)


def show_numbers(*nums):
    print(nums)


def show_details(**info):
    print(info)


greet()
greet_person("Ibrahim")
print(add(5, 10))
welcome()
welcome("Ibrahim")
student_info(age=20, name="Ibrahim")
show_numbers(1, 2, 3, 4)
show_details(name="Ibrahim", age=20)


# VARIABLE SCOPE
x = 10

def test():
    global x
    x = 20
    print("Inside:", x)

test()
print("Outside:", x)


# FUNCTION ANNOTATIONS
def multiply(a: int, b: int) -> int:
    return a * b

print(multiply(4, 5))


# PACKING & UNPACKING
a, b, c = (10, 20, 30)
print(a, b, c)

numbers = [1, 2, 3, 4, 5]
x, *y = numbers
print(x)
print(y)


# BUILT-IN FUNCTIONS
print(len("python"))
print(max(3, 7, 2))
print(min(3, 7, 2))
print(sum([1, 2, 3]))
print(type(10))


# STRING PRACTICE
name = "Ibrahim"
age = 20
print("My name is " + name)
print(f"My name is {name} and I am {age}")

text = "Programming"
print(text[:5])
print(text[3:])
print(text[::-1])

print("Hello\nWorld")
print("He said \"Hi\"")

msg = "  hello python world  "
print(msg.strip())
print(msg.upper())
print(msg.replace("python", "programming"))

email = "user123@gmail.com"
print(email[email.index('@') + 1:])
```
